# Рисуем ровное сердечко

подключаем matplotlib для визуализации графиков и тригонометрические функции из модуля math

```python
from matplotlib import pyplot as plt
from math import sin, cos
```

>$
\begin{cases}
x=16{sin(t)}^{3} & \\
y=13cos(t)-5cos(2t)-2cos(3t)-cos(4t)
\end{cases}
$

```python
x_coord = lambda t: 16 * sin(t)**3
y_coord = lambda t: 13*cos(t) - 5*cos(2*t) - 2*cos(3*t) - cos(4*t)
```

Cоздаем наборы точек для построения

```python
t_tpl = tuple(map(lambda t: t / 100, range(0, 630)))
x_tpl = tuple(map(x_coord, t_tpl))
y_tpl = tuple(map(y_coord, t_tpl))
```

Настраиваем вид графика и строим

```python
fig, ax = plt.subplots()
fig.set_figwidth(12)
fig.set_figheight(12)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.plot(x_tpl, y_tpl, color='red')
```
